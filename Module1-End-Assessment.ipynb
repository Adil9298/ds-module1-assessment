```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ABC-Company-Assesment.csv')
df
```

```python
df.isnull().sum()
```

```python
df['College'] = df['College'].fillna('Unknown')
```

```python
df['Salary'] = pd.to_numeric(df['Salary'])
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Height'] = np.random.randint(150, 181, size=len(df))
df
```

#  employees across each team and its visualization

```python
team_counts = df['Team'].value_counts()
team_counts
```

```python
team_percentage = (team_counts / len(df)) * 100
team_percentage
```

```python
team_distribution = pd.DataFrame({
    'Employee Count': team_counts,
    'Percentage': team_percentage.round(2)
})
print(team_distribution)
```

```python
team_distribution['Percentage'].plot.pie(
    labels=team_distribution.index,
    autopct='%1.1f%%',
    startangle=140,
    figsize=(6, 6),
    title='Employee Distribution by Team'
)
plt.ylabel("")
plt.show()
```

# employees based on their positions and visualization

```python
position_counts = df['Position'].value_counts()
position_distribution = pd.DataFrame({
    'Employee Count': position_counts,
    'Percentage': (position_counts / len(df) * 100).round(2)
})
print(position_distribution)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=position_distribution.index, y=position_distribution['Employee Count'], palette='Set2')
plt.title('Number of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Employee Count')
plt.tight_layout()
plt.show()
```

# predominant age group among employees and visualization

```python
age_points = [15, 20, 25, 30, 35, 40, 45, 50]
age_labels = ['16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50']

df['Age Group'] = pd.cut(df['Age'], bins=age_points, labels=age_labels)

age_group_counts = df['Age Group'].value_counts().sort_index()

print(age_group_counts)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='Set1')
plt.title('Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

# which team and position have the highest salary expenditure with visualization

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum()

position_salary_expenditure = df.groupby('Position')['Salary'].sum()

highest_team_salary = team_salary_expenditure.idxmax(), team_salary_expenditure.max()
highest_position_salary = position_salary_expenditure.idxmax(), position_salary_expenditure.max()

print("Team with highest salary expenditure: ",highest_team_salary)
print("Position with highest salary expenditure: ",highest_position_salary)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=team_salary_expenditure.index, y=team_salary_expenditure.values, palette='Blues')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=position_salary_expenditure.index, y=position_salary_expenditure.values, palette='Purples')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# correlation between age and salary and its visualization

```python
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary: ",correlation)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Salary', color='green')
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Age', y='Salary', scatter_kws={'color': 'green'}, line_kws={'color': 'red'})
plt.title('Age vs Salary with Regression Line')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Age', 'Salary']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix: Age vs Salary')
plt.tight_layout()
plt.show()
```
